```python
dataset.hist(figsize=(14, 12), bins=20, edgecolor='black')
plt.suptitle('Histograms: Distribution of Each Property in the Dataset')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 10))
fig.suptitle('Scatter Plots: Wine Samples And Their Ingredients')

axes = axes.flatten()

colors = ['darkorange', 'blue', 'green', 'red', 'purple', 'brown', 'yellow', 'gray', 'olive', 'teal', 'magenta', 'salmon']

for i, col in enumerate(dataset.columns[:]): 
    axes[i].scatter(dataset.index, dataset[col], color=colors[i], alpha=0.5, label=col)
    axes[i].set_title(col)
    axes[i].set_xlabel('Wine Samples')
    axes[i].set_ylabel(col)
    axes[i].legend()
                    
plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 10))
fig.suptitle('Box Plots: Ingredients Against Wine Quality')

axes = axes.flatten()

for i, col in enumerate(dataset.columns[:-1]): 
    if i < len(axes) - 1:  
        box_colors = plt.cm.viridis(np.linspace(0, 1, len(sorted(dataset['quality'].unique()))))
        box_plot = axes[i].boxplot([dataset[col][dataset['quality'] == q] for q in sorted(dataset['quality'].unique())], patch_artist=True)

        for patch, color in zip(box_plot['boxes'], box_colors):
            patch.set_facecolor(color)

        axes[i].set_title(col)
        axes[i].set_xticklabels(sorted(dataset['quality'].unique()))
        axes[i].set_xlabel('Quality')
        axes[i].set_ylabel(col)

plt.tight_layout(rect=[0, 0, 1, 0.96])

fig.delaxes(axes[-1])

plt.show()
```

```python
plt.hist(dataset['fixed acidity'], bins = 20, edgecolor = 'black')
plt.xlabel('Fixed Acidity')
plt.ylabel('Frequency')
plt.title('Distribution of Fixed Acidity')
plt.show()
```

```python
quality_counts = dataset['quality'].value_counts()
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'gold', 'lightgreen', 'skyblue', 'purple', 'orange'])
plt.title('Proportion of Quality Ratings')
plt.show()
```

```python
plt.scatter(dataset['volatile acidity'], dataset['citric acid'], c='green', alpha=0.5)
plt.title('Volatile Acidity vs Citric Acid')
plt.xlabel('Volatile Acidity')
plt.ylabel('Citric Acid')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.title('Correlation Heatmap')
plt.imshow(dataset.corr(), cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(dataset.corr().columns)), dataset.corr().columns, rotation=90)
plt.yticks(range(len(dataset.corr().columns)), dataset.corr().columns)
plt.show()
```
